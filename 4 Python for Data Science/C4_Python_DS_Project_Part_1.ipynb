```python
# Task 1
body_type_counts = car_df['Body_Type'].value_counts().nlargest(5)
body_type_counts.plot(kind='barh', color='skyblue')
plt.title('Top 5 Most Preferred Car Body Types', fontsize=16)
plt.xlabel('Number of Cars', fontsize=12)
plt.ylabel('Body Type', fontsize=12)
plt.gca().invert_yaxis()  
plt.show()
```

```python
# Task 2
required_modes = ['normal', 'comfort', 'eco', 'sport', 'power']
cars_with_all_modes = car_df[car_df['Drive_Modes'].apply(lambda x: all(mode in str(x).lower() for mode in required_modes))]

print("Cars that have all the specified driving modes:")
print(cars_with_all_modes)

num_cars_with_all_modes = cars_with_all_modes.shape[0]
print(f"\nNumber of cars with all the specified driving modes: {num_cars_with_all_modes}")
```

```python
# Task 3
manufacturer_city_mileage = car_df.groupby('Make')['City_Mileage_km_litre'].mean().sort_values(ascending=False).nlargest(10)
manufacturer_city_mileage.plot(kind='barh', color='lightgreen')
plt.title('Top 10 Car Manufacturers by City Mileage', fontsize=16)
plt.xlabel('Average City Mileage (km/litre)', fontsize=12)
plt.ylabel('Car Manufacturer', fontsize=12)
plt.gca().invert_yaxis()  
plt.show()

print(manufacturer_city_mileage)
```

```python
# Task 4
non_electric_cars = car_df[car_df['Fuel_Type'].str.lower() != 'electric']
top_20_non_electric = non_electric_cars.groupby('Make')['City_Mileage_km_litre'].mean().sort_values(ascending=False).nlargest(20)
top_20_non_electric.plot(kind='barh', color='cornflowerblue')
plt.title('Top 20 Non-Electric Car Manufacturers by City Mileage', fontsize=16)
plt.xlabel('Average City Mileage (km/litre)', fontsize=12)
plt.ylabel('Car Manufacturer', fontsize=12)
plt.gca().invert_yaxis()  # Invert y-axis to show the highest at the top
plt.show()

print(top_20_non_electric)
```

```python
# Task 5
selected_manufacturers = ['Hyundai', 'Mahindra', 'Renault', 'Skoda']
filtered_df = car_df[car_df['Make'].isin(selected_manufacturers)]

mean_city_mileage = pd.pivot_table(filtered_df, values='City_Mileage_km_litre', index='Make', aggfunc='mean')
print("Mean of City Mileage:")
print(mean_city_mileage)

median_city_mileage = pd.pivot_table(filtered_df, values='City_Mileage_km_litre', index='Make', aggfunc='median')
print("\nMedian of City Mileage:")
print(median_city_mileage)

mean_highway_mileage = pd.pivot_table(filtered_df, values='Highway_Mileage_km_litre', index='Make', aggfunc='mean')
print("\nMean of Highway Mileage:")
print(mean_highway_mileage)

median_highway_mileage = pd.pivot_table(filtered_df, values='Highway_Mileage_km_litre', index='Make', aggfunc='median')
print("\nMedian of Highway Mileage:")
print(median_highway_mileage)
```

```python
# Task 6
selected_manufacturers = ['Hyundai', 'Renault', 'Mahindra', 'Skoda']
filtered_df = car_df[car_df['Make'].isin(selected_manufacturers)]

mean_city_mileage = filtered_df.groupby('Make')['City_Mileage_km_litre'].mean()
median_city_mileage = filtered_df.groupby('Make')['City_Mileage_km_litre'].median()
mileage_stats = pd.DataFrame({
    'Mean': mean_city_mileage,
    'Median': median_city_mileage
})

# Plot a stacked column chart
fig, ax = plt.subplots(figsize=(10, 6))
manufacturers = mileage_stats.index
x = np.arange(len(manufacturers))  
width = 0.4  
ax.bar(x, mileage_stats['Mean'], width, label='Mean', color='lightblue')
ax.bar(x, mileage_stats['Median'], width, label='Median', color='lightgreen', bottom=mileage_stats['Mean'])
ax.set_xlabel('Car Manufacturer', fontsize=12)
ax.set_ylabel('City Mileage (km/litre)', fontsize=12)
ax.set_title('Mean and Median City Mileage of Selected Car Manufacturers', fontsize=16)
ax.set_xticks(x)
ax.set_xticklabels(manufacturers)
ax.legend()
plt.show()
```

```python
# Task 7
average_mileage = car_df.groupby('Make')[['City_Mileage_km_litre', 'Highway_Mileage_km_litre']].mean()
average_mileage['Combined_Average_Mileage'] = average_mileage.mean(axis=1)
sorted_mileage = average_mileage['Combined_Average_Mileage'].sort_values()

plt.figure(figsize=(10, 8))
sorted_mileage.plot(kind='barh', color='lightcoral')
plt.title('Average Mileage (City and Highway) of All Manufacturers', fontsize=16)
plt.xlabel('Average Mileage (km/litre)', fontsize=12)
plt.ylabel('Car Manufacturer', fontsize=12)
plt.show()

print(sorted_mileage)
```

```python
# Task 9
selected_columns = ['Valves_Per_Cylinder', 'Kerb_Weight', 'Make', 'Displacement', 
                    'Fuel_Tank_Capacity_litre', 'City_Mileage_km_litre', 
                    'Highway_Mileage_km_litre', 'Power', 'Torque', 'Gross_Vehicle_Weight']
car_df_selected = car_df[selected_columns]
print("Missing values before treatment:")
print(car_df_selected.isnull().sum())

numerical_columns = ['Valves_Per_Cylinder', 'Kerb_Weight', 'Displacement', 
                     'Fuel_Tank_Capacity_litre', 'City_Mileage_km_litre', 
                     'Highway_Mileage_km_litre', 'Power', 'Gross_Vehicle_Weight']

car_df_selected[numerical_columns] = car_df_selected[numerical_columns].fillna(car_df_selected[numerical_columns].mean())
car_df_selected['Torque'] = car_df_selected['Torque'].fillna('Unknown')
car_df_selected['Make'] = car_df_selected['Make'].fillna(car_df_selected['Make'].mode()[0])

print("\nMissing values after treatment:")
print(car_df_selected.isnull().sum())
print("\nCleaned DataFrame:")
print(car_df_selected.head())
```

```python
# Task 10
selected_columns = [
    'Valves_Per_Cylinder', 'Kerb_Weight', 'Make', 'Displacement', 
    'Fuel_Tank_Capacity_litre', 'City_Mileage_km_litre', 
    'Highway_Mileage_km_litre', 'Power', 'Torque', 'Gross_Vehicle_Weight'
]
new_df = car_df[selected_columns]
new_df = new_df.fillna(new_df.mean(numeric_only=True))  # This fills missing values with column mean for numeric data
new_df = new_df.fillna('Unknown')  # This handles non-numeric columns

print(new_df.head())
```

```python
# Task 11

car_df['Kerb_Weight'] = car_df['Kerb_Weight'].astype(str).str.replace('kg', '', regex=False)
car_df['Kerb_Weight'] = pd.to_numeric(car_df['Kerb_Weight'], errors='coerce')  # Convert to numeric, handle errors
car_df['Kerb_Weight'] = car_df['Kerb_Weight'] * 2.2  # Convert kg to pounds
car_df['Gross_Vehicle_Weight'] = car_df['Gross_Vehicle_Weight'].astype(str).str.replace('kg', '', regex=False)
car_df['Gross_Vehicle_Weight'] = pd.to_numeric(car_df['Gross_Vehicle_Weight'], errors='coerce')  # Convert to numeric, handle errors
car_df['Gross_Vehicle_Weight'] = car_df['Gross_Vehicle_Weight'] * 2.2  # Convert kg to pounds

print("Kerb_Weight and Gross_Vehicle_Weight after conversion to pounds:")
print(car_df[['Kerb_Weight', 'Gross_Vehicle_Weight']].head())
```

```python
# Task 12
def coefficient_of_variation(series):
    return series.std() / series.mean()

performance_metrics = car_df.groupby('Make').agg(
    mean_displacement=('Displacement', 'mean'),
    median_displacement=('Displacement', 'median'),
    std_displacement=('Displacement', 'std'),
    cov_displacement=('Displacement', coefficient_of_variation),
    mean_cylinders=('Cylinders', 'mean'),
    median_cylinders=('Cylinders', 'median'),
    std_cylinders=('Cylinders', 'std'),
    cov_cylinders=('Cylinders', coefficient_of_variation),
    mean_valves_per_cylinder=('Valves_Per_Cylinder', 'mean'),
    median_valves_per_cylinder=('Valves_Per_Cylinder', 'median'),
    std_valves_per_cylinder=('Valves_Per_Cylinder', 'std'),
    cov_valves_per_cylinder=('Valves_Per_Cylinder', coefficient_of_variation)
).reset_index()

# Display the calculated metrics
print(performance_metrics)
```

```python
# Task 14
car_df = pd.read_excel(r"C:\Users\AnnamalaiG\OneDrive\Desktop\coures4\DS1_C4_S9_Car_Data.xlsx")
car_df['Kerb_Weight'] = car_df['Kerb_Weight'].astype(str).str.replace('kg', '', regex=False)
car_df['Kerb_Weight'] = pd.to_numeric(car_df['Kerb_Weight'], errors='coerce')
car_df['Gross_Vehicle_Weight'] = car_df['Gross_Vehicle_Weight'].astype(str).str.replace('kg', '', regex=False)
car_df['Gross_Vehicle_Weight'] = pd.to_numeric(car_df['Gross_Vehicle_Weight'], errors='coerce')
weight_stats = car_df.groupby('Make').agg(
    Mean_Kerb_Weight=('Kerb_Weight', 'mean'),
    Median_Kerb_Weight=('Kerb_Weight', 'median'),
    Std_Kerb_Weight=('Kerb_Weight', 'std'),
    Mean_GVW=('Gross_Vehicle_Weight', 'mean'),
    Median_GVW=('Gross_Vehicle_Weight', 'median'),
    Std_GVW=('Gross_Vehicle_Weight', 'std')
)
weight_stats['CV_Kerb_Weight'] = (weight_stats['Std_Kerb_Weight'] / weight_stats['Mean_Kerb_Weight']) * 100
weight_stats['CV_GVW'] = (weight_stats['Std_GVW'] / weight_stats['Mean_GVW']) * 100

print("Performance Comparison Based on Kerb Weight and Gross Vehicle Weight:")
print(weight_stats[['Mean_Kerb_Weight', 'Median_Kerb_Weight', 'Std_Kerb_Weight', 'CV_Kerb_Weight',
                    'Mean_GVW', 'Median_GVW', 'Std_GVW', 'CV_GVW']].round(2))
```
