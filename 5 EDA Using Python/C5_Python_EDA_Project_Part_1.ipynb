```python
# Task 3
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import OneHotEncoder
from sklearn.metrics import classification_report
import statsmodels.api as sm

# Selecting relevant features for analysis
features = ['lead_time', 'adults', 'children', 'babies', 'meal', 'country', 'market_segment', 'is_repeated_guest', 'previous_cancellations', 'previous_bookings_not_canceled']
X = df[features]
y = df['is_canceled']

# Encoding categorical variables
X_encoded = pd.get_dummies(X, drop_first=True)

# Splitting data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X_encoded, y, test_size=0.3, random_state=42)

# Logistic Regression Model
log_model = LogisticRegression(max_iter=1000)
log_model.fit(X_train, y_train)

# Predictions and Classification Report
y_pred = log_model.predict(X_test)
print(classification_report(y_test, y_pred))

# Displaying the coefficients
coefficients = pd.DataFrame(log_model.coef_, columns=X_encoded.columns)
coefficients.loc[0] = log_model.intercept_
coefficients.index = ['Intercept']
print(coefficients)
```

```python
# Task 5
# Convert reservation_status_date to datetime with correct format
df['reservation_status_date'] = pd.to_datetime(df['reservation_status_date'], format='%d-%m-%Y')
# Convert reservation_status_date to datetime, coercing errors
df['reservation_status_date'] = pd.to_datetime(df['reservation_status_date'], format='%d-%m-%Y', errors='coerce')

print(df['reservation_status_date'].head())
print(df['reservation_status_date'].isna().sum())
```

```python

# Assuming df contains the necessary data and the correct columns
df['reservation_status_date'] = pd.to_datetime(df['reservation_status_date'], format='%d-%m-%Y', errors='coerce')
if df['reservation_status_date'].isnull().any():
    print("Some dates could not be converted. Check for invalid date formats.")
df['year'] = df['reservation_status_date'].dt.year
df['month'] = df['reservation_status_date'].dt.month

if 'adr' not in df.columns or 'number_of_rooms' not in df.columns:
    print("Ensure 'adr' and 'number_of_rooms' columns exist in the DataFrame.")
else:
    adr_over_time = df.groupby(['year', 'month']).agg(
        total_revenue=('adr', 'sum'),  # Sum of ADR for the month
        rooms_sold=('number_of_rooms', 'sum')  # Total rooms sold for the month
    ).reset_index(
    adr_over_time['adr'] = adr_over_time['total_revenue'] / adr_over_time['rooms_sold']
    adr_over_time['adr'].fillna(0, inplace=True)
adr_over_time = df.groupby(['year', 'month'])['adr'].mean().reset_index()

# Plotting ADR over time
plt.figure(figsize=(12, 6))
sb.lineplot(x='month', y='adr', hue='year', data=adr_over_time, marker='o')
plt.title('Average Daily Rate (ADR) Over Time')
plt.xlabel('Month')
plt.ylabel('Average Daily Rate (ADR)')
plt.legend(title='Year')
plt.xticks(ticks=range(1, 13), labels=['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'])
plt.grid()
plt.show()
```

```python
# Assuming 'stays_in_week_nights' indicates the number of rooms sold
df['total_revenue'] = df['adr'] * df['stays_in_week_nights'] * (df['adults'] + df['children'] + df['babies'])

# Grouping by market segment to find effective pricing strategies
revenue_analysis = df.groupby('market_segment').agg({
    'total_revenue': 'sum',
    'adr': 'mean',
    'stays_in_week_nights': 'sum'
}).reset_index()

# Sorting by total revenue
revenue_analysis.sort_values(by='total_revenue', ascending=False, inplace=True)
print(revenue_analysis)
```

```python
# Plotting revenue by market segment
plt.figure(figsize=(12, 6))
sb.barplot(x='market_segment', y='total_revenue', data=revenue_analysis)
plt.title('Total Revenue by Market Segment')
plt.xlabel('Market Segment')
plt.ylabel('Total Revenue')
plt.xticks(rotation=45)
plt.show()
```

```python
# Grouping by reserved room type to find average ADR
adr_by_room_type = df.groupby('reserved_room_type')['adr'].mean().reset_index()

# Sorting by ADR
adr_by_room_type.sort_values(by='adr', ascending=False, inplace=True)
print(adr_by_room_type)
```

```python
# Task 6
# Grouping by distribution channel to calculate bookings and cancellations
channel_analysis = df.groupby('distribution_channel').agg({
    'hotel': 'count', 
    'is_canceled': 'mean',  
    'adr': 'mean',  
}).reset_index()
channel_analysis.rename(columns={'hotel': 'total_bookings', 'is_canceled': 'cancellation_rate'}, inplace=True)
channel_analysis['effective_bookings'] = channel_analysis['total_bookings'] * (1 - channel_analysis['cancellation_rate'])
print(channel_analysis)
```
