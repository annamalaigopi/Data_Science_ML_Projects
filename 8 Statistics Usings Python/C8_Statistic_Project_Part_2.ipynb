```python
# Task 1
mu_0 = 221.3    
x_bar = 222.2     
sigma = 74.83    
n = 40            
alpha = 0.05      
z = (x_bar - mu_0) / (sigma / math.sqrt(n))
p_value = 1 - norm.cdf(z)
z_critical = norm.ppf(1 - alpha)
print(f"Z-statistic: {z:.2f}")
print(f"Critical Z-value: {z_critical:.2f}")
print(f"P-value: {p_value:.4f}")
if z > z_critical:
    print("Reject the null hypothesis. The average advertising budget in 2023 is significantly higher.")
else:
    print("Fail to reject the null hypothesis. There is insufficient evidence to claim the average advertising budget in 2023 is higher.")
```

```python

# Task 2
with_cd = df[df['cd'] == 'yes']['price']
without_cd = df[df['cd'] == 'no']['price']
stat, p_levene = levene(with_cd, without_cd)
equal_var = p_levene > 0.05  # If p-value > 0.05, assume equal variances
t_stat, p_value = ttest_ind(with_cd, without_cd, equal_var=equal_var)
print("Levene's Test for Equal Variance:")
print(f"Statistic: {stat:.2f}, P-value: {p_levene:.4f}")
print("")
print("Two-Sample T-Test:")
print(f"T-statistic: {t_stat:.2f}")
print(f"P-value: {p_value:.4f}")
print(f"Equal Variances Assumed: {equal_var}")
print("")
alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis. There is a statistically significant difference in average prices.")
else:
    print("Fail to reject the null hypothesis. There is no statistically significant difference in average prices.")
```

```python
# Task 3
premium_computers = df[df['premium'] == 'yes']['price']
sample_size = 25
sample = premium_computers.sample(n=sample_size, random_state=42)
sample_mean = sample.mean()
sample_std = sample.std()
mu_0 = 2200
t_stat = (sample_mean - mu_0) / (sample_std / np.sqrt(sample_size))
df = sample_size - 1
alpha = 0.05
t_critical = t.ppf(1 - alpha/2, df)
p_value = 2 * (1 - t.cdf(abs(t_stat), df))
print(f"Sample Mean: {sample_mean:.2f}")
print(f"Sample Standard Deviation: {sample_std:.2f}")
print(f"T-statistic: {t_stat:.2f}")
print(f"Critical T-value: {t_critical:.2f}")
print(f"P-value: {p_value:.4f}")
if abs(t_stat) > t_critical:
    print("Reject the null hypothesis. The mean price of premium computers is significantly different from $2200.")
else:
    print("Fail to reject the null hypothesis. There is no significant difference between the mean price of premium computers and $2200.")
```
